```python
#   ,  


def grad_f(r):
    """
      f,  .
    :param r: np.array[2]: float   2
    :return: np.array[2]: float   2
    """
    return np.sin(2 * r)
```

```python
def grad_descent_2d(f, grad_f, lr, num_iter=100, r0=None):
    """
    ,       f   .
        :param f:    
        :param grad_f: ,    f ( ,     grad_f)
        :param lr: learning rate 
        :param num_iter:    
        :return: np.array[num_iter, 2]   (r, f(r))
    """
    if r0 is None:
        r0 = np.random.random(2)

    #       
    #   .    history
    history = []

    #   --   
    curr_r = r0.copy()
    for iter_num in range(num_iter):
        entry = np.hstack((curr_r, f(curr_r)))
        history.append(entry)

        curr_r -= lr * grad_f(curr_r)

    return np.vstack(history)
```

```python
def generate_batches(X, y, batch_size):
    """
    param X: np.array[n_objects, n_features] ---  -
    param y: np.array[n_objects] ---   
    """
    assert len(X) == len(y)
    np.random.seed(42)
    X = np.array(X)
    y = np.array(y)
    perm = np.random.permutation(len(X))
    batch_max = X.shape[0] // batch_size

    for batch_start in range(batch_max):
        batch_idx = perm[batch_start * batch_size : (batch_start + 1) * batch_size]
        X_batch = X[batch_idx]
        y_batch = y[batch_idx]
        yield (X_batch, y_batch)
```

##  3
  ``fit``  ``get_grad``  ``MyLogisticRegression``.


 :

$$Loss(y, p) = -\sum_{i=1}^{\ell} (y_i \log (p_i) + (1 - y_i) \log (1 - p_i))$$


$$ \frac{\partial L}{\partial w} = X^T (p - y)$$


```python
def logit(x, w):
    return np.dot(x, w)


def sigmoid(h):
    return 1.0 / (1 + np.exp(-h))
```

```python
class MyLogisticRegression(object):
    def __init__(self):
        self.w = None

    def fit(self, X, y, epochs=10, lr=0.1, batch_size=100):
        l, n = X.shape
        if self.w is None:
            np.random.seed(42)
            #     
            self.w = np.random.randn(n + 1)

        X_train = np.concatenate((np.ones((l, 1)), X), axis=1)

        losses = []

        #    losses    .      .

        for i in range(epochs):
            for X_batch, y_batch in generate_batches(X_train, y, batch_size):
                #  X_train    1

                predictions = sigmoid(logit(X_batch, self.w))
                #  loss   
                loss = -1 * np.sum(
                    y_batch * np.log(predictions)
                    + (1 - y_batch) * np.log(1 - predictions)
                )

                assert np.array(loss).shape == tuple(), "   !"

                losses.append(loss)

                # YOUR CODE:  self.w    .   self.get_grad   .    learning rate!
                self.w -= lr * self.get_grad(X_batch, y_batch, predictions)

        return losses

    def get_grad(self, X_batch, y_batch, predictions):
        """
        param X_batch: np.array[batch_size, n_features + 1] ---  -
        param y_batch: np.array[batch_size] ---   
        param predictions: np.array[batch_size] ---   

           X_batch     .
             
                
        !     --  BIAS,    .
               :
           ,  .
               @

         -- -     (np.array[n_features + 1])
        """

        #    
        #   

        grad_basic = X_batch.T @ (predictions - y_batch)
        assert grad_basic.shape == (X_batch.shape[1],), (
            "     k_features + 1 "
        )

        return grad_basic

    def predict_proba(self, X):
        l, n = X.shape
        X_ = np.concatenate((np.ones((l, 1)), X), axis=1)
        return sigmoid(logit(X_, self.w))

    def _predict_proba_internal(self, X):
        """
        ,       predict_proba, 
        predict_proba      ,    
            
        """
        return sigmoid(logit(X, self.w))

    def predict(self, X, threshold=0.5):
        return self.predict_proba(X) >= threshold

    def get_weights(self):
        return self.w.copy()
        # copy   .  copy  ,  get_weights()
        #    , , ,    
        # get_weights(),     self.w.     ,
        # (,  fit),  self.w

    def __loss(self, y, p):
        p = np.clip(p, 1e-10, 1 - 1e-10)
        return -np.sum(y * np.log(p) + (1 - y) * np.log(1 - p))
```

##  4
  ``get_grad()``  ``MyElasticLogisticRegression``

:

$$L_1 (w) = \alpha \sum_{j=1}^{n}|w_j| $$

$$L_2 (w) = \beta\sum_{j=1}^{n}w_j^2$$

$$\frac{\partial L_1}{\partial w_1} = \alpha \cdot \mathrm{sign}(w_1)$$

$$ \frac{\partial L_2}{\partial w_1} = 2\beta w_1$$


```python
class MyElasticLogisticRegression(MyLogisticRegression):
    def __init__(self, l1_coef, l2_coef):
        self.l1_coef = l1_coef
        self.l2_coef = l2_coef
        self.w = None

    def get_grad(self, X_batch, y_batch, predictions):
        """
           X_batch     .
               
                 +  
        !     --  BIAS,    .
        Bias     .        :
           ,  .

         -- -     (np.array[n_features + 1])
        """

        grad_basic = (
            X_batch.T @ (predictions - y_batch)
        )  # YOUR CODE:    .  bias-     !

        grad_l1 = (
            self.l1_coef * np.sign(self.w)
        )  # YOUR CODE:    l_1- (     bias)
        grad_l1[0] = 0
        grad_l2 = (
            2 * self.l2_coef * self.w
        )  # YOUR CODE:    l_2- (     bias)
        grad_l2[0] = 0
        #  bias-    !

        assert grad_l1[0] == grad_l2[0] == 0, (
            "Bias     !"
        )
        assert (
            grad_basic.shape == grad_l1.shape == grad_l2.shape == (X_batch.shape[1],)
        ), "     k_features + 1 "

        return grad_basic + grad_l1 + grad_l2
```
